一

首先讲了SGD（随机梯度下降）方法的缺点

如图所示，如果loss在一个方向上梯度变化很快，一个方向上变化缓慢，那么如图可看到在二维上进展就非常缓慢，而神经网络肯定不是二维，那么出现的问题就会非常多。



这是SGD的第二个缺点：如果SGD进入局部最小值，梯度将停留在这个点，因为梯度是零；在鞍点，梯度将为零，所以会将卡住。
同样，在鞍点，梯度为零，也会卡住。



这是第三个缺点：SGD对噪声很敏感，而小批次是有很多噪声的。

然后介绍了一种SGD+动量的方法

这个动量一般取0.9或0.99；能够解决SGD被鞍点和局部最小值困住的缺点。但是他会超过最小值，然后在返回到最小值点。


这是尼瑟夫动量方法：他相比于之前的方法，他的优点在于不会超出最小值再回来，而是直接奔向最小值，但是比SGD+动量要慢

这是Adam法，是目前效果最好的一种优化方法。它结合了SGD+动量+RMSprop来作为梯度更新，使用了固定的偏差来修复一开始梯度可能为0的情况；然后beta1 = 0.9和beta2 = 0.999和learning_rate = 1e-3或5e-4对很多模型来说是个很好的起点！

二



学习率衰减

在迭代过程中可以使用学习率衰变的技术，使得一开始训练速度快来逼近最小值，然后学习率变小，来一点点接近最小值。

图中是两个可供选择的衰减的策略；需要注意的一点是学习衰变在SGD+动量中很常见，但在ADAM中不常见；还有就是不要从一开始就使用学习衰退来选择超参数，先试一试，看看是否需要衰变。

三

加入正则化来防止模型过拟合

正则化是为了减小模型的复杂度，防止模型对训练数据过度拟合，减小测试准确率和真实准确率之间的差距。

另一种比较有效防止过拟合的方法是Dropout，

        即设置一个超参数n（0<n<1），然后随机的使得百分比n的神经元设置为0。
        这个n一般都设置为0.5

这样做的好处是这一整个模型相当于是多个小模型组合起来了，能够提高模型的泛化能力。
