物体检测：与二的不同之处是物体检测要检测一个或模式不同的对象和位置

1.RNN

过程：1.先通过算法筛选出大约2000个ROI（感兴趣区域）；2.所有区域缩放到相同的大小，然后进如卷积神经网络；3.SVM进行最后的线性分类；4.进行矩形框的调整

问题：1.速度慢；2.缩放的想法不行

2.fast R-CNN

它与RCNN主要的区别就是，一开始直接将图片送到卷积神经网络中去，卷积完之后在经过ROI Pooling 而不是筛选出2000个RPI
