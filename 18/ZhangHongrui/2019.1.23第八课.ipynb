一

CPU与GPU

GPU显卡一开始是为了渲染图形来玩游戏或制作3D媒体而开发的。显卡有NVIDIA与AMD两大厂家，而深度学习选择NVIDIA而不是AMD GPU，因为NVIDIA正在推动研究向前深度学习也使其架构更适合深度学习。

CPU拥有更少的内核，但每个内核的速度更快，功能更强大; 非常适合顺序任务。虽然GPU有更多的核心，但每个核心都慢得; 非常适合并行任务。所以矩阵乘法适合GPU的操作，它具有可以并行完成的MxN独立操作。

编程GPU框架：1.CUDA（仅限NVIDIA）在实践中，我们不会去编写并行代码，而是将使用其他人实施和优化的代码！；2.OPENCL；与CUDA类似，但可以在任何GPU上运行，但是通常较慢，所有深度学习软件都没有太多支持。

在训练的时候如果CPU与GPU通信不好的话，会大大加大运行的时间，所以有以下一些解决办法：1.将所有数据读入RAM（如果可能的话）；2.使用SSD而不是HDD；3.使用多个CPU线程来预取数据！当GPU正在计算时，CPU线程将为您获取数据。


其实很多框架都已经为我们实现了这点，所以不用太深入考虑这些。

五


静态与动态图的区别：


1.使用静态图形，框架可以在运行之前来优化图形（图形其实就是计算图，就是计算的流程）


2.静态：构建图形后，可以将其序列化并运行它，而无需构建图形的代码，可以脱离代码去运行；动态：始终需要代码，然后去运行

3.动态图表循环更容易；静态图中循环更复杂。动态图形应用程序包括：循环网络和递归网络。

4.Caffe2使用静态图形并且可以在python中训练模型也适用于IOS和Android

5.Tensorflow / Caffe2在生产中使用很多，特别是在移动设备上。
