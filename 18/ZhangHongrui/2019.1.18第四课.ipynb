概况的说反向传播：首先是正向计算，计算出每个结点上数的值，然后从后往前求导，利用求导的链式规则可以比较简便的从后往前计算出各个位置的梯度。

这是一个比较简单的具体的例子：
  
  最后面的f对自身导数是1；f对q的导数是z；f对z的导数是q；以此类推可以计算出最后的输出f对每个位置的梯度
  



这个图就是根据链式求导法则对一个计算节点，从后往前一般求梯度的形式



这边是一个比较复杂的例子；由此可以看出直接从前到后计算每个位置的导数是有一定困难的；而从后往前就将复杂的求导简化成一步步简单的求导然后一步步乘起来就好了

从这个例子中可以看出：加法结点 前后的梯度是一致的；乘法的话 就是系数值乘以前面的梯度；幂函数的话就是求导就好了









这个例子就主要是用来形象的讲解取最大值门的梯度；两个数中较大值的梯度为1，较小值的梯度为0，然后再乘前面的梯度就好了。
